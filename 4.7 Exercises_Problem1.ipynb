```python
import numpy as np
from PIL import Image

# 이미지 파일을 엽니다.
image_path = 'C:/Users/kimdanny/window.png'
image = Image.open(image_path)

# 이미지를 NumPy 배열로 변환합니다.
image_array = np.array(image)

# NumPy 배열을 텐서로 사용할 수 있습니다.
image_tensor = image_array

print("이미지 텐서:\n", image_tensor)
```

```python
image_tensor = image_array/255
mean_brightness = np.mean(image_tensor)
print("평균 밝기:", mean_brightness)
```

```python
mean_red = np.mean(image_tensor[:,:,0])
mean_green = np.mean(image_tensor[:,:,1])
mean_blue = np.mean(image_tensor[:,:,2])

print("빨간색 채널 평균:", mean_red)
print("초록색 채널 평균:", mean_green)
print("파란색 채널 평균:", mean_blue)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
plt.axis('off')  # 이미지 주변의 축을 제거합니다.
plt.show()
```
